## 01 — Attendance: Top Art Museums by Annual Footfall
*(2020–21 excluded as COVID outliers. All figures in millions of visitors.)*

**Key data points:**
- The Louvre has been the world's most visited art museum every year in this dataset, though its 2024 figure of 8.74M was marginally below its 2023 peak of 8.9M, partly due to Olympic-related closures and security restrictions.
- Shanghai Museum East opened in stages through 2024 and immediately attracted 4.3M visitors in a truncated first year — a signal of extraordinary latent demand in Asia's cultural tourism market.
- The National Gallery (London) remains the single worst-performing major institution relative to its 2019 baseline, sitting 47% below pre-pandemic attendance despite holding its most-attended temporary exhibition ever (Van Gogh: Poets and Lovers).


```python
# ── Attendance data ───────────────────────────────────────────────────────────
attendance = pd.DataFrame([
    {'museum': 'Louvre',                    'y2019': 9.60, 'y2022': 7.70, 'y2023': 8.90, 'y2024': 8.74},
    {'museum': 'British Museum',            'y2019': 6.24, 'y2022': 4.10, 'y2023': 5.82, 'y2024': 6.48},
    {'museum': 'Vatican Museums',           'y2019': 6.88, 'y2022': 5.10, 'y2023': 6.83, 'y2024': 6.83},
    {'museum': 'Metropolitan Museum',       'y2019': 7.00, 'y2022': 3.41, 'y2023': 5.59, 'y2024': 5.70},
    {'museum': 'Tate Modern',               'y2019': 6.10, 'y2022': 4.70, 'y2023': 4.72, 'y2024': 4.60},
    {'museum': 'Nat. Gallery of Art (DC)',  'y2019': 4.70, 'y2022': 3.00, 'y2023': 4.10, 'y2024': 4.30},
    {'museum': 'Shanghai Museum East',      'y2019': None, 'y2022': None, 'y2023': None,  'y2024': 4.30},
    {'museum': 'Uffizi',                    'y2019': 4.40, 'y2022': 3.80, 'y2023': 4.00, 'y2024': 4.10},
    {'museum': 'Nat. Museum Korea',         'y2019': 3.35, 'y2022': 3.40, 'y2023': 4.18, 'y2024': 4.00},
    {'museum': 'National Gallery (London)', 'y2019': 5.19, 'y2022': 2.70, 'y2023': 3.00, 'y2024': 3.10},
    {'museum': 'MoMA',                      'y2019': 3.50, 'y2022': 2.30, 'y2023': 2.86, 'y2024': 2.70},
    {'museum': 'Louvre Abu Dhabi',          'y2019': 1.00, 'y2022': 0.90, 'y2023': 1.20, 'y2024': 1.40},
])
attendance.set_index('museum', inplace=True)
print(attendance.to_string())
```

## 02 — Recovery vs 2019 Baseline
**How much of each institution's pre-pandemic footfall has been reclaimed?**

The recovery split is blunt and revealing. Institutions with state backing or strong domestic catchments recovered fastest. UK institutions dependent on international tourism — particularly Tate Modern and the National Gallery — are the worst performers, and the data now suggests some of this deficit may be structural rather than temporary.


```python
# ── Recovery analysis ─────────────────────────────────────────────────────────
rec = attendance.dropna(subset=['y2019', 'y2024']).copy()
rec['pct_recovery'] = (rec['y2024'] / rec['y2019'] * 100).round(1)
rec['shortfall_M']  = (rec['y2019'] - rec['y2024']).round(2)
rec = rec.sort_values('pct_recovery', ascending=True)

def recovery_colour(pct):
    if pct >= 100: return GREEN
    if pct >= 85:  return ACCENT
    return RED

colours = [recovery_colour(p) for p in rec['pct_recovery']]

fig, ax = plt.subplots(figsize=(11, 5))
fig.patch.set_facecolor(DARK)
ax.set_facecolor(MID)

bars = ax.barh(rec.index, rec['pct_recovery'], color=colours, zorder=3, height=0.7)
ax.axvline(100, color=ACCENT, linestyle='--', linewidth=1.2, zorder=4)
ax.text(101, len(rec) - 0.5, '2019 level', color=ACCENT, fontsize=8, va='top')

for bar, val in zip(bars, rec['pct_recovery']):
    ax.text(bar.get_width() + 0.8, bar.get_y() + bar.get_height() / 2,
            f'{val:.0f}%', va='center', fontsize=8.5,
            color=recovery_colour(val), fontweight='bold')

ax.set_xlabel('% of 2019 attendance achieved in 2024', fontsize=9)
ax.set_title('Post-pandemic recovery vs 2019 baseline  |  Green ≥ 100%  |  Amber ≥ 85%  |  Red < 85%', pad=10)
ax.set_xlim(0, 125)
ax.grid(axis='x', zorder=0)
ax.spines[:].set_visible(False)

plt.tight_layout()
plt.savefig('chart_recovery.png', dpi=150, bbox_inches='tight', facecolor=DARK)
plt.show()

print(rec[['pct_recovery', 'shortfall_M']].rename(
    columns={'pct_recovery': '% of 2019', 'shortfall_M': 'Visitor shortfall (M)'}
).to_string())
```

## 03 — Funding & Financial Resilience

There are three distinct funding worlds in this analysis, not two:

1. **US endowment model** — the Met ($3.3B), Getty ($8.4B), MoMA (~$1.5B). Permanent financial independence. Largely insulated from political cycles.
2. **Gulf sovereign wealth model** — Louvre Abu Dhabi and Qatar Museums. Effectively unlimited state backing. No financial ceiling on ambition.
3. **UK mixed-model** — Tate, British Museum, National Gallery. Dependent on a combination of DCMS grants (down 32% in real terms since 2010), earned income, and corporate sponsorship. Structurally precarious.

European continental institutions (Louvre, Uffizi) benefit from more stable state support and sit between models 1 and 3.

**The DCMS funding erosion is material.** Government spending per UK resident on culture now stands at £32.79 versus £48.43 in 2010/11 — a 32.3% real-terms fall, per Campaign for the Arts analysis published May 2025. DCMS's budget for 2025/26 is 5.9% below the previous year in real terms, even as total government departmental spending grew 5% in the same period.


```python
# ── Funding data ──────────────────────────────────────────────────────────────
funding = pd.DataFrame([
    {'institution': 'Getty Trust',            'endowment_usd_bn': 8.4,  'annual_state_usd_bn': 0.00, 'model': 'US endowment'},
    {'institution': 'Met Museum',             'endowment_usd_bn': 3.3,  'annual_state_usd_bn': 0.00, 'model': 'US endowment'},
    {'institution': 'MoMA',                   'endowment_usd_bn': 1.5,  'annual_state_usd_bn': 0.00, 'model': 'US endowment'},
    {'institution': 'Louvre Abu Dhabi',       'endowment_usd_bn': 0.00, 'annual_state_usd_bn': 15.0, 'model': 'Gulf sovereign'},
    {'institution': 'Qatar Museums',          'endowment_usd_bn': 0.00, 'annual_state_usd_bn': 12.0, 'model': 'Gulf sovereign'},
    {'institution': 'Louvre (France)',        'endowment_usd_bn': 0.10, 'annual_state_usd_bn': 0.50, 'model': 'European state'},
    {'institution': 'British Museum',         'endowment_usd_bn': 0.02, 'annual_state_usd_bn': 0.07, 'model': 'UK mixed'},
    {'institution': 'Tate (all sites)',       'endowment_usd_bn': 0.05, 'annual_state_usd_bn': 0.06, 'model': 'UK mixed'},
    {'institution': 'National Gallery (Lon)', 'endowment_usd_bn': 0.01, 'annual_state_usd_bn': 0.05, 'model': 'UK mixed'},
])

model_colours = {
    'US endowment':    GREEN,
    'Gulf sovereign':  ORANGE,
    'European state':  BLUE,
    'UK mixed':        RED,
}

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.patch.set_facecolor(DARK)
for ax in axes:
    ax.set_facecolor(MID)
    ax.spines[:].set_visible(False)
    ax.grid(axis='x', zorder=0)

# Left: Endowment
df_e = funding.sort_values('endowment_usd_bn')
cols_e = [model_colours[m] for m in df_e['model']]
axes[0].barh(df_e['institution'], df_e['endowment_usd_bn'], color=cols_e, zorder=3, height=0.65)
axes[0].set_title('Private / Endowment Capital (USD bn)', fontsize=10)
axes[0].set_xlabel('USD billions', fontsize=9)
for val, y in zip(df_e['endowment_usd_bn'], range(len(df_e))):
    if val > 0:
        axes[0].text(val + 0.05, y, f'${val:.1f}B', va='center', fontsize=8,
                     color=model_colours[df_e.iloc[y]['model']])

# Right: Annual state backing
df_s = funding.sort_values('annual_state_usd_bn')
cols_s = [model_colours[m] for m in df_s['model']]
axes[1].barh(df_s['institution'], df_s['annual_state_usd_bn'], color=cols_s, zorder=3, height=0.65)
axes[1].set_title('Annual State / Sovereign Grant Estimate (USD bn)', fontsize=10)
axes[1].set_xlabel('USD billions (estimated)', fontsize=9)
for val, y in zip(df_s['annual_state_usd_bn'], range(len(df_s))):
    if val > 0:
        axes[1].text(val + 0.1, y, f'${val:.2f}B', va='center', fontsize=8,
                     color=model_colours[df_s.iloc[y]['model']])

# Legend
patches = [mpatches.Patch(color=c, label=m) for m, c in model_colours.items()]
fig.legend(handles=patches, loc='lower center', ncol=4, frameon=True,
           facecolor=MID, edgecolor='#444', fontsize=9, bbox_to_anchor=(0.5, -0.02))

plt.suptitle('Museum Funding: Three Distinct Models', fontsize=12, color=LIGHT, y=1.01)
plt.tight_layout()
plt.savefig('chart_funding.png', dpi=150, bbox_inches='tight', facecolor=DARK)
plt.show()
print("Gulf sovereign figures are estimates based on published infrastructure commitments and cultural district budgets.")
print("UK state figures are DCMS grant-in-aid as reported in respective 2023–24 annual reports.")
```

## 04 — Global Tourism Trends & Projected Museum Impact

Global tourism fully recovered to 2019 levels in 2024 and set a new post-pandemic record in 2025 with 1.52 billion international arrivals. UN Tourism forecasts 3–4% annual growth through 2026, with major boosts expected from the FIFA World Cup 2026 (US/Canada/Mexico) and Milano Cortina Winter Olympics.

**Regional divergence is critical for museum analysis:**
- **Europe:** +4% in 2025 vs 2024; +6% above 2019. This is the bedrock of visitor volume for the Louvre, British Museum, Uffizi and Vatican.
- **Asia Pacific:** +11% in H1 2025. Japan (+21%), South Korea (+15%), Vietnam (+21%). The cultural tourism market from this region is growing fast and these visitors disproportionately attend art museums.
- **Middle East:** -4% in H1 2025. The only region in decline. Discussed in depth in Section 6.
- **Americas:** Flat (+0% in North America H1 2025). Softening US outbound travel is a headwind for European museums.


```python
# ── Global tourism trends and projection ──────────────────────────────────────
tourism = pd.DataFrame([
    {'year': '2019',  'arrivals_bn': 1.465, 'actual': True},
    {'year': '2022',  'arrivals_bn': 0.963, 'actual': True},
    {'year': '2023',  'arrivals_bn': 1.286, 'actual': True},
    {'year': '2024',  'arrivals_bn': 1.470, 'actual': True},
    {'year': '2025',  'arrivals_bn': 1.520, 'actual': True},
    {'year': '2026e', 'arrivals_bn': 1.572, 'actual': False},
    {'year': '2027e', 'arrivals_bn': 1.620, 'actual': False},
    {'year': '2028e', 'arrivals_bn': 1.668, 'actual': False},
    {'year': '2029e', 'arrivals_bn': 1.718, 'actual': False},
    {'year': '2030e', 'arrivals_bn': 1.800, 'actual': False},
])

fig, ax = plt.subplots(figsize=(12, 5))
fig.patch.set_facecolor(DARK)
ax.set_facecolor(MID)
ax.spines[:].set_visible(False)
ax.grid(axis='y', zorder=0)

actual    = tourism[tourism['actual']]
projected = tourism[~tourism['actual']]

# Connect actual to first projected point
bridge = pd.concat([actual.tail(1), projected])

ax.plot(actual['year'],    actual['arrivals_bn'] * 1000,
        color=ACCENT, linewidth=2.5, marker='o', markersize=6, zorder=4, label='Actual')
ax.plot(bridge['year'],    bridge['arrivals_bn'] * 1000,
        color=ACCENT, linewidth=2, linestyle='--', marker='o', markersize=5,
        zorder=4, alpha=0.7, label='Projected (UNWTO 3–4% pa)')

ax.axvline('2025', color='#555', linewidth=1, linestyle=':', zorder=3)
ax.text('2025', ax.get_ylim()[0] + 50, '← Actual  |  Projected →',
        fontsize=8, color=FADED, ha='center')

ax.set_ylabel('International arrivals (millions)', fontsize=9)
ax.set_title('Global International Tourist Arrivals 2019–2030e  |  COVID years excluded', pad=10)
ax.legend(loc='upper left')

for _, row in tourism.iterrows():
    ax.annotate(f"{row['arrivals_bn']*1000:.0f}M",
                xy=(row['year'], row['arrivals_bn'] * 1000),
                xytext=(0, 10), textcoords='offset points',
                ha='center', fontsize=7.5,
                color=ACCENT if row['actual'] else FADED)

plt.tight_layout()
plt.savefig('chart_tourism.png', dpi=150, bbox_inches='tight', facecolor=DARK)
plt.show()
print("Sources: UN Tourism World Tourism Barometer (January 2026). 2026–2030 projected at 3% pa (conservative UNWTO range).")
```

## 05 — Projected Attendance Trajectories: 2024–2030

Projections are modelled on three inputs: (a) current recovery trajectory from 2019 baseline, (b) institution-specific funding outlook and its effect on programming quality, and (c) share of inbound tourism in the relevant destination city.

**Louvre Abu Dhabi** is shown with a downside-risk band reflecting regional geopolitical uncertainty (see Section 6). The base case assumes relative stability. The bear case assumes sustained conflict-driven decline in Western leisure tourism to the Gulf.

**Tate Modern** is the only institution in this analysis projected to be meaningfully smaller in 2030 than it is today — a consequence of structural funding erosion, post-pandemic behavioural shifts, and growing competition from within London's own museum ecosystem.

**National Gallery (London)** shows a recovery inflection point in 2025–26, driven by the Sainsbury Wing reopening. Whether this inflection is sustained depends heavily on whether UK museums can attract back the European tourists who have not fully returned since Brexit.


```python
# ── Projected attendance model ─────────────────────────────────────────────────
proj_years = [2019, 2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030]

institutions = {
    'Louvre':                   [9.60, 7.70, 8.90, 8.74, 8.85, 9.00, 9.15, 9.30, 9.50, 9.70],
    'British Museum':           [6.24, 4.10, 5.82, 6.48, 6.70, 6.90, 7.05, 7.15, 7.25, 7.35],
    'Vatican Museums':          [6.88, 5.10, 6.83, 6.83, 6.90, 6.95, 7.00, 7.00, 7.00, 7.00],
    'Metropolitan Museum':      [7.00, 3.41, 5.59, 5.70, 5.90, 6.10, 6.30, 6.50, 6.65, 6.80],
    'Tate Modern':              [6.10, 4.70, 4.72, 4.60, 4.50, 4.45, 4.40, 4.35, 4.30, 4.25],
    'Nat. Gallery (London)':    [5.19, 2.70, 3.00, 3.10, 3.50, 3.90, 4.10, 4.20, 4.30, 4.40],
    'Louvre Abu Dhabi (base)':  [1.00, 0.90, 1.20, 1.40, 1.55, 1.72, 2.00, 2.35, 2.65, 2.90],
    'Louvre Abu Dhabi (bear)':  [1.00, 0.90, 1.20, 1.40, 1.35, 1.30, 1.30, 1.35, 1.40, 1.50],
}

palette = [ACCENT, BLUE, PURPLE, RED, '#e74c3c', TEAL, ORANGE, ORANGE]

fig, ax = plt.subplots(figsize=(14, 7))
fig.patch.set_facecolor(DARK)
ax.set_facecolor(MID)
ax.spines[:].set_visible(False)
ax.grid(axis='y', zorder=0, alpha=0.5)

for (name, vals), colour in zip(institutions.items(), palette):
    ls = '--' if 'bear' in name else '-'
    lw = 1.5 if 'bear' in name else 2.2
    alpha = 0.6 if 'bear' in name else 1.0
    label = name if 'bear' not in name else 'Louvre Abu Dhabi (bear case)'
    ax.plot(proj_years, vals, color=colour, linewidth=lw,
            linestyle=ls, alpha=alpha, label=label, marker='o', markersize=3.5, zorder=3)

# Shade Abu Dhabi uncertainty band
lad_base = institutions['Louvre Abu Dhabi (base)']
lad_bear = institutions['Louvre Abu Dhabi (bear)']
ax.fill_between(proj_years, lad_bear, lad_base,
                color=ORANGE, alpha=0.12, label='Abu Dhabi risk band', zorder=2)

ax.axvline(2024.5, color='#555', linewidth=1, linestyle=':', zorder=4)
ax.text(2024.7, 0.3, '← Actual  |  Projected →', fontsize=8, color=FADED)

ax.set_xticks(proj_years)
ax.set_xticklabels([str(y) for y in proj_years], fontsize=9)
ax.set_ylabel('Annual visitors (millions)', fontsize=9)
ax.set_title('Projected Attendance Trajectories 2019–2030  |  COVID years excluded from trend lines', pad=10)
ax.legend(loc='upper left', ncol=2, fontsize=8)
ax.set_ylim(0, 11)

plt.tight_layout()
plt.savefig('chart_projections.png', dpi=150, bbox_inches='tight', facecolor=DARK)
plt.show()
print("Projections are modelled estimates, not forecasts. Carry meaningful uncertainty particularly beyond 2027.")
print("Louvre Abu Dhabi bear case assumes sustained geopolitical risk suppresses Western leisure tourism to the Gulf.")
```

## 06 — The Middle East: A Disrupted Ambition

This section requires more than a chart. The situation has evolved materially and rapidly.

**The trajectory through 2024** was, for Gulf institutions, broadly positive. Despite the Gaza war beginning in October 2023 — which caused the Travel Sentiment Score for the entire Middle East region to fall to near-pandemic lows according to World Bank data — the UAE and Qatar demonstrated real resilience. Qatar's inbound tourism grew 25% year-on-year in 2024, reaching 5 million international arrivals. Louvre Abu Dhabi recorded its best attendance ever at 1.4 million visitors. The divergence within the region was sharp: non-GCC Middle Eastern economies (Jordan, Lebanon) suffered material declines; the Gulf states held firm.

**What changed in 2025:**

- **June 2025:** The United States and Israel conducted military strikes on Iranian nuclear facilities. Airspace closures, travel advisories, and flight cancellations followed across the region. The Middle East recorded -4% tourist arrivals in H1 2025 — the only region globally in decline.

- **September 2025:** Israel struck Hamas political leadership meeting in Doha, killing six people including a Qatari security officer and damaging buildings in the Qatari capital. This was the first direct military strike by any foreign power on a Gulf state in modern history. UAE President Sheikh Mohamed bin Zayed Al Nahyan condemned it as "criminal" and flew to Doha within 24 hours. A UN Security Council statement condemning the strike was adopted unanimously.

**What this means for the museums:**

For **Qatar's institutions** (Museum of Islamic Art, National Museum of Qatar), the immediate operational impact has been limited. But Qatar's entire cultural-diplomatic positioning rests on projecting an image of stability, neutrality and bridge-building. That image is now significantly strained. Qatar's role as a Hamas mediator — previously a source of international respectability — has become a liability in terms of the physical security risks it attracted.

For **Louvre Abu Dhabi**, the picture is more structurally complex. The museum's visitor base is 84% international, with China and Russia leading at 12% each, followed by India (7%), France and the UK (6% each). Western visitors are not the majority, which provides some insulation from Western political sentiment. However, the UAE's maintenance of Abraham Accords ties with Israel has been deeply unpopular across the Arab world: a January 2024 poll found 67% of respondents across sixteen Arabic-speaking countries considered the UAE's approach "bad or very bad." Emirati citizens have reportedly faced harassment when travelling in the region. The museum's stated mission of universal cultural dialogue sits in some tension with its government's geopolitical positioning.

**The Saadiyat Cultural District** expansion — which is to add a Guggenheim Abu Dhabi, Zayed National Museum, and a Natural History Museum to the existing Louvre Abu Dhabi — remains in planning. The Guggenheim Abu Dhabi has been in planning for nearly two decades. The pace of delivery will be read, internationally, as a signal of how confident the UAE remains in its soft-power cultural strategy.

## 07 — Beyond Footfall: Qualitative Importance

Attendance is the most measurable proxy for cultural importance, but it is an incomplete one. The Vatican Museums would be one of the world's most important institutions even if they closed to the public tomorrow. MoMA defines contemporary art discourse to a degree that far outweighs its visitor numbers. The Louvre Abu Dhabi's acquisition budget is transforming the global supply of great Islamic and non-Western art.

The following six dimensions are rated 1–10 for each institution. Weights reflect the relative importance of each factor to long-run cultural significance:

| Dimension | Weight | Rationale |
|---|---|---|
| Masterpiece density | 1.5 | Irreplaceable works are the foundation of long-run importance |
| Institutional prestige | 1.5 | Peer recognition, scholar access, global brand |
| Acquisitions / programme | 1.2 | Current investment in future relevance |
| Growth trajectory | 1.2 | Forward momentum in attendance and influence |
| Press coverage | 1.0 | Media footprint; cultural agenda-setting power |
| Digital reach | 0.8 | Instagram, online engagement; expanding beyond physical walls |


```python
# ── Qualitative scores and composite ranking ──────────────────────────────────
qual = pd.DataFrame([
    {'institution': 'Metropolitan Museum',      'masterpieces': 9, 'prestige': 10, 'acquisitions': 9, 'growth': 7, 'press': 9,  'digital': 9},
    {'institution': 'Louvre',                   'masterpieces': 10,'prestige': 10, 'acquisitions': 7, 'growth': 6, 'press': 9,  'digital': 8},
    {'institution': 'MoMA',                     'masterpieces': 8, 'prestige': 9,  'acquisitions': 8, 'growth': 5, 'press': 9,  'digital': 10},
    {'institution': 'British Museum',           'masterpieces': 9, 'prestige': 9,  'acquisitions': 7, 'growth': 8, 'press': 8,  'digital': 7},
    {'institution': 'Uffizi',                   'masterpieces': 10,'prestige': 9,  'acquisitions': 5, 'growth': 5, 'press': 7,  'digital': 7},
    {'institution': 'Vatican Museums',          'masterpieces': 10,'prestige': 10, 'acquisitions': 4, 'growth': 4, 'press': 7,  'digital': 5},
    {'institution': 'National Gallery (London)','masterpieces': 9, 'prestige': 9,  'acquisitions': 6, 'growth': 6, 'press': 7,  'digital': 7},
    {'institution': 'Tate Modern',              'masterpieces': 7, 'prestige': 8,  'acquisitions': 6, 'growth': 3, 'press': 8,  'digital': 8},
    {'institution': 'Louvre Abu Dhabi',         'masterpieces': 6, 'prestige': 7,  'acquisitions': 9, 'growth': 9, 'press': 7,  'digital': 6},
    {'institution': 'Qatar: MIA / NMoQ',        'masterpieces': 7, 'prestige': 7,  'acquisitions': 9, 'growth': 6, 'press': 6,  'digital': 5},
    {'institution': 'Nat. Museum Korea',        'masterpieces': 7, 'prestige': 7,  'acquisitions': 6, 'growth': 8, 'press': 6,  'digital': 7},
    {'institution': 'Nat. Gallery of Art (DC)', 'masterpieces': 8, 'prestige': 8,  'acquisitions': 7, 'growth': 6, 'press': 7,  'digital': 7},
])

weights = {'masterpieces': 1.5, 'prestige': 1.5, 'acquisitions': 1.2,
           'growth': 1.2, 'press': 1.0, 'digital': 0.8}
total_weight = sum(weights.values())

def composite(row):
    return sum(row[k] * v for k, v in weights.items()) / total_weight

qual['score_now'] = qual.apply(composite, axis=1).round(2)

# Future adjustment: modify growth score ±
future_adj = {
    'Metropolitan Museum': +0.3, 'Louvre': 0, 'MoMA': -0.7,
    'British Museum': +0.5, 'Uffizi': +0.2, 'Vatican Museums': -0.5,
    'National Gallery (London)': +0.5, 'Tate Modern': -1.5,
    'Louvre Abu Dhabi': +1.5, 'Qatar: MIA / NMoQ': +0.3,
    'Nat. Museum Korea': +0.8, 'Nat. Gallery of Art (DC)': +0.2,
}

def score_future(row):
    adj = future_adj.get(row['institution'], 0)
    r2 = row.copy()
    r2['growth'] = min(10, max(1, row['growth'] + adj))
    return composite(r2)

qual['score_2030'] = qual.apply(score_future, axis=1).round(2)
qual = qual.sort_values('score_now', ascending=False).reset_index(drop=True)
qual['rank_now']  = qual['score_now'].rank(ascending=False).astype(int)
qual['rank_2030'] = qual['score_2030'].rank(ascending=False).astype(int)
qual['rank_change'] = qual['rank_now'] - qual['rank_2030']  # positive = risen

print(qual[['institution','score_now','score_2030','rank_now','rank_2030','rank_change']].to_string(index=False))
```

```python
# ── Composite score chart ─────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(12, 6))
fig.patch.set_facecolor(DARK)
ax.set_facecolor(MID)
ax.spines[:].set_visible(False)
ax.grid(axis='x', zorder=0, alpha=0.5)

y = np.arange(len(qual))
w = 0.38

bars_now = ax.barh(y + w/2, qual['score_now'],  height=w, color=ACCENT,         label='Composite score now',   zorder=3)
bars_fut = ax.barh(y - w/2, qual['score_2030'], height=w, color=ACCENT + '66',  label='Projected score ~2030', zorder=3)

ax.set_yticks(y)
ax.set_yticklabels(qual['institution'], fontsize=9)
ax.set_xlabel('Weighted composite score (max 10)', fontsize=9)
ax.set_title('Qualitative Importance: Composite Score Now vs Projected 2030', pad=10)
ax.legend(loc='lower right')
ax.set_xlim(0, 10.5)

# Annotate rank change arrows
for i, row in qual.iterrows():
    change = row['rank_change']
    if change != 0:
        colour = GREEN if change > 0 else RED
        symbol = f'↑{abs(change)}' if change > 0 else f'↓{abs(change)}'
        ax.text(max(row['score_now'], row['score_2030']) + 0.15, i,
                symbol, va='center', fontsize=8.5, color=colour, fontweight='bold')

plt.tight_layout()
plt.savefig('chart_qualitative.png', dpi=150, bbox_inches='tight', facecolor=DARK)
plt.show()
```

## 08 — Radar Profiles: Individual Institution Analysis

Radar charts for five contrasting institutions. These give a quick read of each institution's strengths and weaknesses across the six qualitative dimensions.


```python
# ── Radar charts for five selected institutions ────────────────────────────────
from matplotlib.patches import FancyArrowPatch

selected = ['Metropolitan Museum', 'Louvre', 'Tate Modern',
            'Louvre Abu Dhabi', 'Vatican Museums']

dims = ['masterpieces', 'digital', 'press', 'acquisitions', 'prestige', 'growth']
dim_labels = ['Masterpieces', 'Digital', 'Press', 'Acquisitions', 'Prestige', 'Growth']
n = len(dims)
angles = np.linspace(0, 2 * np.pi, n, endpoint=False).tolist()
angles += angles[:1]  # close the polygon

colours_r = [ACCENT, BLUE, RED, ORANGE, PURPLE]

fig, axes = plt.subplots(1, 5, figsize=(18, 4), subplot_kw=dict(polar=True))
fig.patch.set_facecolor(DARK)

for ax, name, colour in zip(axes, selected, colours_r):
    ax.set_facecolor(MID)
    row = qual[qual['institution'] == name].iloc[0]
    vals = [row[d] for d in dims] + [row[dims[0]]]

    ax.plot(angles, vals, color=colour, linewidth=2)
    ax.fill(angles, vals, color=colour, alpha=0.2)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(dim_labels, fontsize=7.5, color=LIGHT)
    ax.set_yticks([2, 4, 6, 8, 10])
    ax.set_yticklabels(['2','4','6','8','10'], fontsize=6, color=FADED)
    ax.set_ylim(0, 10)
    ax.set_title(name, fontsize=9, color=colour, pad=10, fontweight='bold')
    ax.tick_params(colors=FADED)
    ax.spines['polar'].set_color('#444')
    ax.grid(color='#444', linewidth=0.5)

    score = row['score_now']
    ax.text(0, -3.5, f'Score: {score:.2f}', ha='center', fontsize=8,
            color=colour, transform=ax.transData)

plt.suptitle('Qualitative Dimension Profiles — Selected Institutions', fontsize=11, color=LIGHT, y=1.04)
plt.tight_layout()
plt.savefig('chart_radar.png', dpi=150, bbox_inches='tight', facecolor=DARK)
plt.show()
```

### Conclusions

The data makes one thing unmistakeably clear: the pandemic did not treat all great institutions equally, and neither has the recovery. Museums backed by sovereign wealth — Louvre Abu Dhabi, Qatar's institutions, and to a lesser extent the great French national collections — are operating in a fundamentally different financial reality to their UK counterparts. While Tate and the National Gallery worry about government grant allocations and deficit budgets, Abu Dhabi is planning a cultural district that will eventually house multiple global-calibre institutions within walking distance of each other. The comparison is not flattering for Britain.

That said, money alone does not make a great art institution. The Metropolitan Museum's supremacy in this analysis rests not just on its $3.3 billion endowment but on the irreplaceable depth of a collection built over 150 years, the quality of its curatorial staff, and a digital strategy that has turned it into a genuinely global cultural force. The Louvre's dominance is inseparable from the Mona Lisa and from Paris — two things no amount of Gulf spending can replicate. Institutional prestige accretes slowly and its foundations are deep.

The most important structural trend in this analysis is the divergence between the Western and Gulf models, accelerated by the ongoing regional conflict. If the ceasefire holds, if Gulf security stabilises, and if the Saadiyat Cultural District opens broadly on schedule, then Louvre Abu Dhabi becomes a genuinely different kind of institution by 2030 — one that combines Western collection depth with Gulf financial power and Asian visitor demographics. China and Russia currently account for 24% of its international visitors, and India for a further 7%. No institution in art history has ever had access to that particular combination of resources and visitor demographics simultaneously.

The risks, however, are not trivial. The September 2025 Israeli strike on Doha was a threshold event. A continuation of the conflict, or sustained reputational concerns about the UAE's political positioning among the Western cultural community, could meaningfully slow the Gulf's cultural ambitions. These institutions are spending billions acquiring great art. They may find the harder challenge is acquiring the trust of the international cultural community — a currency that cannot simply be bought.

In the UK, the picture is one of managed decline dressed up as strategic renewal. The Tate Future Fund is necessary but at £43 million raised so far will take until the late 2030s to materially change Tate's funding structure. British Museum stands apart: visitor recovery is real, its collection extraordinary, and the redevelopment programme gives it a medium-term story. But even the British Museum is one bad government decision away from a funding crisis.

**The five most important art institutions in 2030, in order:** Metropolitan Museum · Louvre · British Museum · Louvre Abu Dhabi (geopolitical risk caveat) · Vatican Museums. Tate Modern will still matter, but less so. If Shanghai Museum East achieves in its second and third full years what it achieved in its first, Asia will demand a more prominent position in this conversation far sooner than most Western analysts currently anticipate.
